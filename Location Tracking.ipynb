# Above and Beyond: Trajectory Analysis!

I will plot the trip on the map for July 4th, 2015. Upon investigating the data points, we observed that movement occurred between 2:20 PM and 4:15 PM. However, examining each point individually is time-consuming. Instead, letâ€™s explore how the object moved by reviewing key points along the trajectory.




```python
# Convert index (timestamp) to a column
july_4_2015_data.reset_index(inplace=True)

# Ensure the 'timestamp' column is in proper datetime format
july_4_2015_data['timestamp'] = pd.to_datetime(july_4_2015_data['timestamp'])

# Filter the data for the time range between 2:20 PM and 4:15 PM
filtered_data = july_4_2015_data[(july_4_2015_data['timestamp'].dt.time >= pd.to_datetime('14:20').time()) &
                                 (july_4_2015_data['timestamp'].dt.time <= pd.to_datetime('16:15').time())]


# Creates a map with animation from 2:20PM : 4:15PM
fig = px.scatter_mapbox(filtered_data,
                            lat="latitude",
                            lon="longitude",
                            hover_name="timestamp",
                            hover_data={"latitude": True, "longitude": True},
                            animation_frame=filtered_data['timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S'),
                            zoom=10,
                            height=600)


fig.update_layout(mapbox_style="open-street-map", title='Trajectory Analysis from 2:20PM to 4:15PM')
fig.show()
```

This map shows the entire trajectory of the object's movement from 2:20 PM to 4:15 PM. The trip occurred in a counterclockwise direction, starting near the northeastern side of Walla Walla. The object followed a loop-like pattern, indicating a systematic coverage of that region.

Notably, at 3:23:39 PM, there was an unusual deviation where the object appears significantly farther away from the main trajectory, moving rapidly toward the highway (WA 125). This sudden shift suggests a possible GPS error, as the movement speed required for such a distance in that short period seems unlikely. Despite this anomaly, the overall path is consistent and smooth, showing a controlled movement across the map
